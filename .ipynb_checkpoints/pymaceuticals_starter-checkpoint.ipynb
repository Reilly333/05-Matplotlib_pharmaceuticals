```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

drug_list= ["Capomulin", "Infubinol", "Ketapril", "Placebo"]

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
clinical_data = pd.read_csv(clinical_trial_data_to_load)
# Combine the data into a single dataset
merged_df =mouse_data.merge(clinical_data)


# Display the data table for preview

merged_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor_vol = (merged_df.groupby(['Drug', 'Timepoint']).mean()['Tumor Volume (mm3)'])
# Convert to DataFrame
mean_tumor_vol_df = pd.DataFrame(mean_tumor_vol).reset_index()
# Preview DataFrame
mean_tumor_vol_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_vol_standard_err = (merged_df.groupby(['Drug','Timepoint']).sem()['Tumor Volume (mm3)'])
# Convert to DataFrame
tumor_vol_standard_err_df = pd.DataFrame(tumor_vol_standard_err).reset_index
# Preview DataFrame
tumor_vol_standard_err_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_tumor_vol_grouped_df =mean_tumor_vol_df.set_index(['Timepoint','Drug']).unstack()
# Preview that Reformatting worked
mean_tumor_vol_grouped_df
```

```python
# Generate the Plot (with Error Bars)
plt.plot(mean_tumor_vol_grouped_df.loc[:,("Tumor Volume (mm3)",['Capomulin'])], color="red", marker="o",label="Capomulin",linestyle="dotted",linewidth=1)
plt.plot(mean_tumor_vol_grouped_df.loc[:,("Tumor Volume (mm3)",['Infubinol'])], color="blue",marker="^",label="Infubinol",linestyle="dotted",linewidth=1)
plt.plot(mean_tumor_vol_grouped_df.loc[:,("Tumor Volume (mm3)",['Ketapril'])], color="green",marker="s",label="Ketapril",linestyle="dotted",linewidth=1)
plt.plot(mean_tumor_vol_grouped_df.loc[:,("Tumor Volume (mm3)",['Placebo'])], color="black",marker="d",label="Placebo",linestyle="dotted",linewidth=1)
plt.legend()
#error bars

standard_error_timepoint =tumor_vol_standard_err_df[tumor_vol_standard_err_df["Drug"].isin(drug_list)]["Timepoint"]
mean_error_vol=mean_tumor_vol_df[mean_tumor_vol_df["Drug"].isin(drug_list)]["Tumor Volume (mm3)"]
error_vol_2=tumor_vol_standard_err_df[tumor_vol_standard_err_df["Drug"].isin(drug_list)]["Tumor Volume (mm3)"]

plt.errorbar(standard_error,mean_error_vol,mean_error_vol=error_vol_2,fmt='|',color="black")


plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume(mm3)")
plt.grid()
# Save the Figure


```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
group_count_df = pd.DataFrame(merged_df.groupby(['Drug', 'Timepoint']).count()['Mouse ID'])
group_count_df.reset_index(inplace=True)

# Convert to DataFrame

# Preview DataFrame
group_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
group_count_reformat=group_count_df.set_index(["Timepoint","Drug"]).unstack()
# Preview the Data Frame
group_count_reformat
```

```python
# Generate the Plot (Accounting for percentages)
group_count_reformat ["Mouse ID"] = 100*group_count_reformat["Mouse ID"]/(group_count_reformat["Mouse ID"].max().max())
plt.plot(group_count_reformat.loc[:,("Mouse ID",['Capomulin'])], color="red", marker="o",label="Capomulin",linestyle="dotted",linewidth=1)
plt.plot(group_count_reformat.loc[:,("Mouse ID",['Infubinol'])], color="blue",marker="^",label="Infubinol",linestyle="dotted",linewidth=1)
plt.plot(group_count_reformat.loc[:,("Mouse ID",['Ketapril'])], color="green",marker="s",label="Ketapril",linestyle="dotted",linewidth=1)
plt.plot(group_count_reformat.loc[:,("Mouse ID",['Placebo'])], color="black",marker="d",label="Placebo",linestyle="dotted",linewidth=1)                                                                         
#plt ledgend 
plt.legend()

#plt labels
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.grid()

# Save the Figure

# Show the Figure
plt.show()
```
